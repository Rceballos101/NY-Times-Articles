# Preprocessing / Cleaning the data
Remove Duplicates/Missing data and articles outside of time period manually in excel

```python
import pandas
```

```python
data = pandas.read_excel('articles.xlsx' )
articles = data.copy()
articles.tail()
```

```python
print('shape',articles.shape)

print(articles.columns)
```

Reorganize column placement

```python
order = ['id', 'news_desk', 'pub_date', 'section', 'source', 'type', 'url', 'word_count', 'locations', 'subjects', 'headline', 'snippet',]
articles = articles.reindex(columns = order)
articles.tail()
```

```python
articles.snippet[3]
```

Merge headline column and snippet

```python
headlineSnippet = []

for i in range(len(articles.snippet)):
    headlineSnippet.append(articles.headline[i] + ' - ' + articles.snippet[i])
# merge headline and snippet together
headlineSnippet[1:3]
```

```python
headlineSnippet1 = pandas.DataFrame({'headlineSnippet' : headlineSnippet})
len(headlineSnippet1) == len(articles.snippet)

articles = articles.join(headlineSnippet1) # merge it to existing dataframe
articles.head()
```

```python
articles.headlineSnippet[50000]
```

# Remove unnecessary articles

```python
junk = ['Admin','Arts','Books','Briefing','Corrections','Crosswords & Games','Fashion & Style','Food','Movies',
        'Multimedia','Multimedia/Photos','NA','NYT Now','Paid Death Notices','Podcasts','Style','Theater','Todays Paper' ]
# articles to be removed if they have the above section. They don't seem to add much value to our topic and would distract us
articles.shape
```

```python
for i in junk: # len(articles.section)
    articles = articles[articles.section != i] #filter out articles in junk

articles.shape
```

```python
news = ['National','Washington','Politics','U.S.','U.S. / Politics'] # ,'Editorial','Blogs','OpEd','Opinion'

articles = articles[articles.news_desk.isin(news)]
len(articles)
```

```python
set(articles.section) # remaining sections ok as they must be important
```

```python
articles.index = [i for i in range(len(articles))]
articles.index

articles.tail()
```

# Tokenize

```python
import nltk
```

```python
tokens = [] # list of tokenized articles

for i in articles.headlineSnippet:
    tokens.append(' '.join(nltk.word_tokenize(i)))
    
tokens[-10:]
```

```python
tokensDf = pandas.DataFrame({'tokens' : tokens})
tokensDf.tail()
```

```python
print(len(tokensDf) == len(articles.snippet))

articles = articles.join(tokensDf)
articles.tail()
```

# Punctuation and lower case

```python
punctuations = ".,\"-\\/#!?$%\^&\*;:{}=\-_'~()‘’“”_/--'`''\\_@<>|"
tokensPunc = [] # list of punctuation removed

for i in articles.tokens:
    art = []
    for j in i.split():
        if j not in punctuations:
            art.append(''.join(j.lower()))
    tokensPunc.append(' '.join(art))
tokensPunc[:5]
```

```python
tokensDf1 = pandas.DataFrame({'tokensNoPunc' : tokensPunc})
tokensDf1.tail()
```

```python
print(len(tokensDf1) == len(articles.snippet))

articles = articles.join(tokensDf1)
articles.tail()
```

```python
articles.tokensNoPunc[0]
```

```python
s = set()
for i in articles.tokensNoPunc:
    for j in i.split():
        s.add(j)

s= list(s)
[i for i in sorted(s, key = len) if len(i) < 3]
# use to check what punctuation is remaining
```

# Stopwords

```python
from nltk.corpus import stopwords
stopword = stopwords.words('english')

stopword # words to be removed
```

```python
filtered = [] # list of punctuation removed
for i in articles.tokensNoPunc:
    art = []
    for j in i.split():
        if j not in stopword and len(j) > 2:
            art.append(''.join(j))
    filtered.append(' '.join(art))
filtered[:5]
```

```python
tokensDf2 = pandas.DataFrame({'tokensFiltered' : filtered})
tokensDf2.tail()
```

```python
print(len(tokensDf2) == len(articles.snippet))

articles = articles.join(tokensDf2)
articles.tail()
```

```python
articles.tokensFiltered[19000]
```

```python
s = set()
for i in articles.tokensFiltered:
    for j in i.split():
        s.add(j)
s= list(s)
[i for i in sorted(s, key = len) if len(i) < 4]
# use to check what short words are remaining that may need to be removed
```

# Parts of Speech

```python
articles.tokensFiltered[0]
```

```python
pos = []
for i in articles.tokensFiltered:
    if i == "":
        pos.append([('null','null')])
    else:
        pos.append(nltk.pos_tag(i.split()))
```

```python
pos1 = []
for i in pos:
    tag = []
    for j in i:
        tag.append(j[-1])
    pos1.append(' '.join(tag))
    
postag = pandas.DataFrame({'postag' : pos1})
postag.tail()
```

```python
print(len(articles.snippet) == len(postag))

articles = articles.join(postag)
articles.tail()
```

# Stem

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = []
for i in articles.tokensFiltered:
    snip = []
    for j in i.split():
        snip.append(PorterStemmer().stem(j))
    stemmer.append(' '.join(snip))
```

```python
stemmer = pandas.DataFrame({'stem' : stemmer})
stemmer.tail()
```

```python
print(len(articles.snippet) == len(stemmer))

articles = articles.join(stemmer)
articles.tail()
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma = []
for i in articles.tokensFiltered:
    l = []
    for j in i.split():
        l.append(WordNetLemmatizer().lemmatize(j))
    lemma.append(' '.join(l))
```

```python
lemma = pandas.DataFrame({'lemma' : lemma})
lemma.tail()
```

```python
print(len(articles.snippet) == len(lemma))

articles = articles.join(lemma)
articles.tail()
```

```python
articles = articles[articles.tokensFiltered != '']
articles.index = [i for i in range(len(articles))]
articles.tail()
```
